# JS website scraping with selenium

```python
import urllib.request
from bs4 import BeautifulSoup
import json
import requests
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import os
```

```python
#launch url
url = "http://kanview.ks.gov/PayRates/PayRates_Agency.aspx"

# create session
driver = webdriver.Chrome('D:\Coursera\ProjectX\chromedriver.exe')
driver.implicitly_wait(30)
driver.get(url)

soup_lv0=BeautifulSoup(driver.page_source, 'lxml')
title_lv0 = soup_lv0.findAll('a', id = re.compile('^MainContent_uxLevel1_Agencies_uxAgencyBtn_'))

datalist = []
idx_1 = 1
for tt0 in title_lv0:
    
    # set limit to our scraping results
    if idx_1 >= 5:
        driver.quit()
        break
    
    
    
    python_button = driver.find_element_by_id(tt0['id'])
    python_button.click() #click fhsu link

    soup_lv1=BeautifulSoup(driver.page_source, 'lxml')
    title_lv1 = soup_lv1.findAll('a', id = re.compile('^MainContent_uxLevel2_JobTitles_uxJobTitleBtn_'))
    
    print(idx_1, '/', len(title_lv0), 'Level 1: ID: ', tt0['id'])
    
    idx_2 = 1
    
    for tt1 in title_lv1:
        python_button = driver.find_element_by_id(tt1['id'])
        python_button.click() #click fhsu link
        soup_lv2=BeautifulSoup(driver.page_source, 'lxml')
        
        # store data
        table = soup_lv2.find_all('table')[0]
        df = pd.read_html(str(table),header=0)
        datalist.append(df[0])
    
        #click the back button
        driver.execute_script("window.history.go(-1)") 
        print('    ', idx_2, '/', len(title_lv1), 'Level 2: ID: ', tt1['id'])
        idx_2 += 1
    
        
    driver.execute_script("window.history.go(-1)")
    idx_1 += 1

driver.quit()
```
