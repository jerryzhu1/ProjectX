```python
import urllib.request
from bs4 import BeautifulSoup
import json
import requests
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import os
import time
```

```python
driver = webdriver.Chrome('D:\Coursera\ProjectX\chromedriver.exe')
driver.implicitly_wait(10)

url_lv1 = []

for i in range(19):
    url = "https://www.cryptocompare.com/coins/list/USD/" + str(i+1)
    
    driver.get(url)
    
    soup_lv0=BeautifulSoup(driver.page_source, 'lxml')
    
    title_lv0 = soup_lv0.findAll('a', title = re.compile('See full markets and forum for'))
    
    for i in title_lv0:
        if ('https://www.cryptocompare.com' + i['href']) not in url_lv1:
            url_lv1.append('https://www.cryptocompare.com' + i['href'])

    
#driver.close()
```

```python
df = pd.DataFrame([])
url_lv1 = url_lv1[0:5]
for url in url_lv1:
    print(url)
    driver.get(url)
    
    soup_lv1=BeautifulSoup(driver.page_source, 'lxml')
    
    python_buttons = driver.find_elements_by_tag_name('md-tab-item')
    
    i =0
    for python_button in python_buttons:
        time.sleep(2)
        python_button.click() #click fhsu link
        
        if i == 0:
            test1 = soup_lv1.find_all('div', class_='info-text ng-binding')
            table1 = []
            for table_con in test1:
                if table_con.text != '':
                    table1.append(table_con.text)

            table_need = table1[0:11]
            df1 = pd.DataFrame(table_need).T
        i+=1 

        try:
            test2 = soup_lv1.find('div', class_='coin-description ng-binding')
            df2 = pd.DataFrame([test2.text])
            df1 = df1.join(df2.rename(columns={0:'fact %(d)s' %{"d":i}}))
        except:
            pass
        print('Running: tab %(d)s' %{"d":i})
    df = df.append(df1)
        
```
